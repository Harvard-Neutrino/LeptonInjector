```python
import sys
import os
import os.path
import numpy as np
import h5py as h5
import json
from tqdm import tqdm
import argparse
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
data_file = h5.File("test.h5",'r')


injector_list = [i for i in data_file.keys()]
props = None
mu_props = None
injector_n = []
for i in injector_list:
    p = data_file[i]["properties"][:]
    p.dtype.names = (
            'energy',
            'zenith',
            'azimuth',
            'bjorken_x',
            'bjorken_y',
            'final_type_0',
            'final_type_1',
            'particle',
            'x',
            'y',
            'z',
            'total_column_depth'
            )
    mp = data_file[i]['final_1'][:]
    mp.dtype.names = ('_', 'particle', 'position', 'direction', 'energy')
    injector_n.append(len(p))
    if props is None:
        props = p
    else:
        props = np.concatenate([props, p])
    if mu_props is None:
        mu_props = mp
    else:
        mu_props = np.concatenate([mu_props, mp])
```

```python
data = {}
for i,k in enumerate(p.dtype.names):
    data[k] = [_p[i] for _p in p]
```

```python
for i,k in enumerate(p.dtype.names):
    if k=='energy': 
        plt.hist(data[k],bins=10,range=(0,3000))
        plt.semilogy()
    else: plt.hist(data[k],bins=10)
    plt.title(k)
    plt.show()
```

```python
plt.hist2d(data['x'],data['y'],bins=20)
plt.show()

plt.hist2d(data['x'],data['z'],bins=20)
plt.show()

plt.hist2d(data['y'],data['z'],bins=20)
plt.show()

fig = plt.figure()
ax = plt.axes(projection="3d")

ax.scatter3D(data['x'], data['y'], data['z'], cmap='hsv');

plt.show()
```
